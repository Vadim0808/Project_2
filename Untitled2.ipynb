```python
# Removing all rows with NaN values in the CustomerID column because we cannot cluster them without knowing the client's ID. And then we look at how much data remains.
df = df.dropna(subset=['CustomerID'])
df['CustomerID'].isnull().sum(), df.shape
```

```python
df = df[df['Country'] == 'United Kingdom'] # Saving data for users from the United Kingdom only, as the representation from other countries is too small.
df.shape
```

```python
df['Date'] = pd.to_datetime(df['InvoiceDate']).dt.date   # Creating a new column 'Date' in the format of 'YYYY-MM-DD' (e.g., 2021-10-24) for more accurate processing
df.head()
```

```python
# Since these are old data and we cannot measure how long ago a customer made a purchase, we take the data of the last transaction and will measure relative to it
df['Date'] = pd.to_datetime(df['InvoiceDate']).dt.date
NOW = df['Date'].max()
NOW
```

```python
#Next, we find the day of the last purchase by the customer and set it as the variable 'Last_Purchase_Date',
# and also calculate the number of days since the customer's last purchase into the variable 'Recency'
R_df = df.groupby('CustomerID').agg({'Date': 'max'}).reset_index()
R_df.columns = ['CustomerID', 'Last_Purchase_Date']
R_df['Recency'] = (NOW - R_df['Last_Purchase_Date']).apply(lambda x: x.days)
R_df.head()
```

```python
#Here, I determine 'Frequency' - the number of customer purchases (the number of unique invoices)
#'Monetary' - how much money the customer spent in the store
frequency = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
frequency.columns = ['CustomerID', 'Frequency']
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
monetary = df.groupby('CustomerID')['TotalPrice'].sum().reset_index()
monetary.columns = ['CustomerID', 'Monetary']
FM_df = frequency.merge(monetary, on='CustomerID')
FM_df.head()
```

```python
#Now, we combine the necessary variables into 'RFM_df'
RFM_df = R_df.merge(FM_df, on='CustomerID')
RFM_df.head()
```

```python
# Constructing a histogram with the distribution of data and also calculating the coefficient of variation. We see that the data are heterogeneous with a large deviation.
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(RFM_df['Recency'], kde=True, color="skyblue")
plt.title('Recency Distribution')
plt.subplot(1, 3, 2)
sns.histplot(RFM_df['Frequency'], kde=True, color="olive")
plt.title('Frequency Distribution')
plt.subplot(1, 3, 3)
sns.histplot(RFM_df['Monetary'], kde=True, color="gold")
plt.title('Monetary Distribution')
plt.tight_layout()
plt.show()
recency_skewness = RFM_df['Recency'].skew()
frequency_skewness = RFM_df['Frequency'].skew()
monetary_skewness = RFM_df['Monetary'].skew()
print("Recency Skewness: ", recency_skewness)
print("Frequency Skewness: ", frequency_skewness)
print("Monetary Skewness: ", monetary_skewness)
```

```python
#To reduce skewness, we apply the transformation log(x+1) to the data
RFM_df['Recency_log'] = np.log1p(RFM_df['Recency'])
RFM_df['Frequency_log'] = np.log1p(RFM_df['Frequency'])
RFM_df['Monetary_log'] = np.log1p(RFM_df['Monetary'])
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(RFM_df['Recency_log'], kde=True, color="skyblue")
plt.title('Log Transformed Recency Distribution')
plt.subplot(1, 3, 2)
sns.histplot(RFM_df['Frequency_log'], kde=True, color="olive")
plt.title('Log Transformed Frequency Distribution')

plt.subplot(1, 3, 3)
sns.histplot(RFM_df['Monetary_log'], kde=True, color="gold")
plt.title('Log Transformed Monetary Distribution')

plt.tight_layout()
plt.show()
recency_log_skewness = RFM_df['Recency_log'].skew()
frequency_log_skewness = RFM_df['Frequency_log'].skew()
monetary_log_skewness = RFM_df['Monetary_log'].skew()

recency_log_skewness, frequency_log_skewness, monetary_log_skewness
```

```python
# Since there is still skewness, we correct it using the technique of Winsorizing.
windsoriser = Winsorizer(tail='both', fold=2, variables=['Recency_log', 'Frequency_log', 'Monetary_log', 'Recency', 'Frequency', 'Monetary'])
RFM_df[['Recency_log', 'Frequency_log', 'Monetary_log']].isnull().sum()
RFM_df['Monetary_log'].fillna(1, inplace=True)
windsoriser.fit(RFM_df)
RFM_df_winsorised = windsoriser.transform(RFM_df)
```

```python
# Standardizing the values of interest and examining the statistics for them
scaler = StandardScaler()
columns_to_scale = ['Recency_log', 'Frequency_log', 'Monetary_log']
RFM_df[columns_to_scale] = scaler.fit_transform(RFM_df[columns_to_scale])
RFM_df[columns_to_scale].describe()
```

```python
#Using the Elbow Method to determine the optimal number of clusters. Based on the graph, the optimal number of clusters is between 3 and 5.
columns_to_cluster = ['Recency_log', 'Frequency_log', 'Monetary_log']
data_to_cluster = RFM_df[columns_to_cluster]
inertia_values = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_to_cluster)
    inertia_values.append(kmeans.inertia_)
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia_values, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
#Visualizing the clusters obtained using K-means
data = RFM_df[['Recency_log', 'Frequency_log', 'Monetary_log']]
def fit_visualize_kmeans(data, clusters_number, original_df_rfm):
    kmeans = KMeans(n_clusters=clusters_number, random_state=42)
    kmeans.fit(data)
    cluster_labels = kmeans.labels_
    df_new = original_df_rfm.copy()
    df_new['Cluster'] = cluster_labels
    model = TSNE(random_state=1)
    transformed = model.fit_transform(df_new[['Recency', 'Frequency', 'Monetary']])
    plt.title('Flattened Graph of {} Clusters'.format(clusters_number))
    params = dict(
        x=transformed[:, 0],
        y=transformed[:, 1],
        hue='Cluster',
        style='Cluster',
        palette="Set1",
        data=df_new
    )
    sns.scatterplot(**params)
    return df_new
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(3, 1, 1)
df_rfm_k3 = fit_visualize_kmeans(data, 3, RFM_df)
plt.subplot(3, 1, 2)
df_rfm_k4 = fit_visualize_kmeans(data, 4, RFM_df)
plt.subplot(3, 1, 3)
df_rfm_k5 = fit_visualize_kmeans(data, 5, RFM_df)
plt.tight_layout()
```

```python
#Additionally, analyzing the data using Snake plots to better understand the distribution relative to the variables
data_to_cluster = RFM_df[['Recency_log', 'Frequency_log', 'Monetary_log']]
kmeans_3 = KMeans(n_clusters=3, random_state=42).fit(data_to_cluster)
kmeans_4 = KMeans(n_clusters=4, random_state=42).fit(data_to_cluster)
kmeans_5 = KMeans(n_clusters=5, random_state=42).fit(data_to_cluster)
RFM_df['Cluster_3'] = kmeans_3.labels_
RFM_df['Cluster_4'] = kmeans_4.labels_
RFM_df['Cluster_5'] = kmeans_5.labels_
def snake_plot(RFM_df, cluster_labels, n_clusters):
    normalised_df_rfm = RFM_df[['CustomerID', 'Recency_log', 'Frequency_log', 'Monetary_log']].copy()
    normalised_df_rfm['Cluster'] = cluster_labels
    df_melt = pd.melt(normalised_df_rfm.reset_index(),
                        id_vars=['CustomerID', 'Cluster'],
                        value_vars=['Recency_log', 'Frequency_log', 'Monetary_log'],
                        var_name='Metric',
                        value_name='Value')
    plt.xlabel('Metric')
    plt.ylabel('Value')
    sns.pointplot(data=df_melt, x='Metric', y='Value', hue='Cluster')
```

```python
plt.figure(figsize=(9, 9))
plt.subplot(3, 1, 1)
plt.title('Snake Plot of K-Means = 3')
snake_plot(RFM_df, RFM_df['Cluster_3'], 3)
plt.subplot(3, 1, 2)
plt.title('Snake Plot of K-Means = 4')
snake_plot(RFM_df, RFM_df['Cluster_4'], 4)
plt.subplot(3, 1, 3)
plt.title('Snake Plot of K-Means = 5')
snake_plot(RFM_df, RFM_df['Cluster_5'], 5)
plt.tight_layout()
```

```python
#For the final analysis, I choose to distribute into 4 groups and create a table with target variables, adding a variable with the percentage of product returns by group for a deeper analysis.
prod_return = df.groupby('CustomerID').apply(lambda x: (x['Quantity'] < 0).sum()).reset_index(name='NegativeQuantityCount')
RFM_df = RFM_df.merge(prod_return, on='CustomerID', how='left').fillna(0)
RFM_df['NegativeQuantityCount'] = RFM_df['NegativeQuantityCount'].fillna(0).astype(int)
cluster_aggregate = RFM_df.groupby('Cluster_4').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean',
    'CustomerID': 'count',
    'NegativeQuantityCount': 'sum'
}).rename(columns={'CustomerID': 'CustomerPercentage'})
total_customers = RFM_df['CustomerID'].nunique()
cluster_aggregate['CustomerPercentage'] = (cluster_aggregate['CustomerPercentage'] / total_customers * 100).round(2).astype(str) + '%'
total_returns = RFM_df['NegativeQuantityCount'].sum()
cluster_aggregate['ReturnPercentage'] = (cluster_aggregate['NegativeQuantityCount'] / total_returns * 100).round(2).astype(str) + '%'
cluster_aggregate.drop(columns=['NegativeQuantityCount'], inplace=True)
cluster_aggregate
```

Analyzing the outcomes of the four identified customer groups reveals distinct behaviors and potential strategies for engagement:

 The first group I would call "Potential Regular Customers" because these clients have made their purchases recently and have not yet fully shown their
 potential. Also, this group has the lowest percentage of returns, which is good, indicating that some of them may be one-time customers who might not return.

 The second group I would call "Regular Customers." This group is the most active; they make purchases more frequently and for the largest amounts. They are, on one hand, the most valuable customers. However, it's also important to note the highest percentage of returns, despite being the smallest group. This could indicate that they are the most demanding customers, and it might be beneficial to analyze the products they purchase and possibly improve their quality. Despite the returns, customers continue to come back. But, of course, it would be possible to analyze this group more closely and divide it into subgroups to identify even more valuable customers.

 The third group I would call "Cold Customers." This group, mostly, has not made purchases for a long time, even though it is the least numerous. Most likely, these are mostly one-time customers, and the small percentage of returns, despite the number, may indicate just quick one-time purchases. This is more likely a group of one-time customers and those who make rare and inexpensive purchases. Many of them will churn.

 The fourth group I would call "Prospective Customers." This group is the second largest in number and is almost as valuable as the "Regular Customers," despite making purchases less frequently and on average cheaper, they are more numerous. The frequency and cost of their purchases put them in second place, making them promising. It would be advisable to stimulate them, and they also have the second highest rate of returns, which also indicates their engagement, and this should be considered, as with the second group.

 In summary, the second and fourth groups are currently the customer backbone of the store, the main buyers. The first group is recent buyers, and they may either drop out or become part of the backbone. The third group is the most passive and numerous group, and most likely some customers will not return or will make rare and numerous purchases.
